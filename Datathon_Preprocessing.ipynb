```python
from scipy.stats import linregress
from matplotlib import pyplot as plt


# Select the two columns for which you want to plot the line of best fit
column1 = 'Percent below poverty line'
column2 = 'Family median income'

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(FLcounty_statistics[column1], FLcounty_statistics[column2])

# Generate x values for the line of best fit
x_values = np.linspace(min(FLcounty_statistics[column1]), max(FLcounty_statistics[column1]), 100)

# Calculate corresponding y values using the equation of the line
y_values = slope * x_values + intercept

# Plot the scatter plot
plt.scatter(FLcounty_statistics[column1], FLcounty_statistics[column2], label='Data')

# Plot the line of best fit
plt.plot(x_values, y_values, color='red', label=f'Line of Best Fit (r={r_value:.2f})')

# Add labels and title
plt.xlabel(column1)
plt.ylabel(column2)
plt.title(f'Line of Best Fit between {column1} and {column2}')
plt.legend()
plt.grid(True)
plt.show()
```

# **Add "At Risk" Column**

```python
# Calculate the threshold values for poverty and median income rank
poverty_threshold = 12
income_rank_threshold = 104000

# Define a function to determine if a row is at risk
def is_at_risk(row):
    if row['Coastal Flooding Risk Score'] >= 80 and row['Percent below poverty line'] >= poverty_threshold and row['Family median income'] <= income_rank_threshold:
        return 1
    else:
        return 0

# Apply the function to create the "At Risk" column
FLcounty_statistics['At Risk'] = FLcounty_statistics.apply(is_at_risk, axis=1)

# Print the DataFrame to verify the changes
FLcounty_statistics
```

```python
# Apply the function to create the "At Risk" column
GAcounty_statistics['At Risk'] = GAcounty_statistics.apply(is_at_risk, axis=1)

# Print the DataFrame to verify the changes
GAcounty_statistics
```

```python
# Apply the function to create the "At Risk" column
TXcounty_statistics['At Risk'] = TXcounty_statistics.apply(is_at_risk, axis=1)

# Print the DataFrame to verify the changes
TXcounty_statistics
```

```python
# Get the counts of each unique category
FLcategory_counts = FLcounty_statistics['At Risk'].value_counts()

# Print the unique categories and their counts
for category, count in FLcategory_counts.items():
    print(f"Florida Risk Category: {category}, Florida Risk Count: {count}")
```

```python
# Get the counts of each unique category
GAcategory_counts = GAcounty_statistics['At Risk'].value_counts()

# Print the unique categories and their counts
for category, count in GAcategory_counts.items():
    print(f"Georgia Risk Category: {category}, Georgia Risk Count: {count}")
```
